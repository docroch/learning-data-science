```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
ct = pd.crosstab(data['Attrition'], data['JobSatisfaction'], margins=True)
```

```python
ct
```

```python
ct.iloc[0] / ct.iloc[0]['All']
```

```python
ct.iloc[1] / ct.iloc[1]['All']
```

```python
ct.iloc[0][: -1]
```

```python
ct.iloc[0] / ct.iloc[0]['All'] - ct.iloc[1] / ct.iloc[1]['All']
```

```python
obs = np.append(ct.iloc[0][: -1].values, ct.iloc[1][: -1].values)
```

```python
obs
```

```python
row_sum = ct.iloc[0: 2, 4].values
row_sum
```

```python
ct.loc['All', 'All']
```

```python
exp = []
for j in range(2):
    for val in ct.iloc[2, :-1].values:
        tmp = val * row_sum[j] / ct.loc['All', 'All']
        print(f"{val} * {row_sum[j]} / {ct.loc['All', 'All']}  = {tmp}")
        exp.append(tmp)
```

```python
exp
```

```python
((obs - exp) ** 2 / exp).sum()
```

```python
dof = (len(row_sum) - 1) * (len(ct.iloc[2, :-1].values) - 1)
```

```python
dof
```

```python
1 - stats.chi2.cdf(((obs - exp) ** 2 / exp).sum(), dof)
```
