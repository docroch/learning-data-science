# Normality Test
Is data normal?

```python
from scipy.stats import shapiro
import numpy as np
```

```python
data = np.random.normal(loc=20, scale=5, size=1000)
data
```

if (pvalue > 0.05) -> normal data

```python
shapiro(data)
```

```python
from scipy import stats
import pylab
```

```python
stats.probplot(data, dist='norm', plot=pylab)
```
