```python
fig, (ax1, ax2) = plt.subplots(2)
ax1.hist(df['comb08'], range=(10, 50))
ax1.set(title='Vehicle MPG', xlabel='MPG', ylabel='Count')

ax2.boxplot(df['comb08'], tick_labels=['All'], vert=False, showfliers=False)
ax2.set(ylabel='MPG for vehicles')

fig.suptitle('MPG Distribution', fontsize=14, fontweight='bold')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9,4))
ax1.hist(df['comb08'], range=(10, 50))
ax1.set(title='Vehicle MPG', xlabel='MPG', ylabel='Count')

ax2.boxplot(df['comb08'], tick_labels=['All'], showfliers=False)
ax2.set(ylabel='MPG for vehicles')

fig.suptitle('MPG Distribution', fontsize=14, fontweight='bold')
```

```python
df.groupby('year')['comb08'].mean().round(2)
```

```python
avg_by_year = df.groupby(['year'], as_index=False).agg({'comb08': 'mean'}).round(2)
avg_by_year
```

```python
fig, ax1 = plt.subplots()
ax1.plot(avg_by_year['year'], avg_by_year['comb08'])
ax1.set(title='Avg By Year', ylabel='MPG', xlabel='Year');
```

```python
fig, ax1 = plt.subplots()
ax1.plot(avg_by_year['year'], avg_by_year['comb08'])
ax1.set_xticks(np.arange(2000, 2022, 2))
ax1.set(title='Avg By Year', ylabel='MPG', xlabel='Year');
```

```python
fig, ax1 = plt.subplots()
ax1.bar(avg_by_year['year'], avg_by_year['comb08'])
ax1.set(title='Avg By Year', ylabel='MPG', xlabel='Year');
```

```python
fig, ax1 = plt.subplots()
ax1.bar(avg_by_year['year'], avg_by_year['comb08'])
ax1.set_xticks(np.arange(2000, 2022, 2))
ax1.set(title='Avg By Year', ylabel='MPG', xlabel='Year');
```

```python
fig, ax1 = plt.subplots()
ax1.barh(avg_by_year['year'], avg_by_year['comb08'])
ax1.set(title='Avg By Year', xlabel='MPG', ylabel='Year');
```

```python
fig, ax1 = plt.subplots()
ax1.barh(avg_by_year['year'], avg_by_year['comb08'])
ax1.set_yticks(np.arange(2000, 2022, 2))
ax1.set(title='Avg By Year', xlabel='MPG', ylabel='Year');
```
