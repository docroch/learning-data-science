```python
avg_by_year = df.groupby(['year'], as_index=False).agg({'comb08': 'mean'}).round(2)
avg_by_year
```

```python
fig, ax1 = plt.subplots()
ax1.plot(avg_by_year['year'], avg_by_year['comb08'])
ax1.set(title='Avg By Year', ylabel='MPG', xlabel='Year');
```

```python
fig, ax1 = plt.subplots()
ax1.plot(avg_by_year['year'], avg_by_year['comb08'])
ax1.set_xticks(np.arange(2000, 2022, 2))
ax1.set(title='Avg By Year', ylabel='MPG', xlabel='Year');
```

```python
fig, ax1 = plt.subplots()
ax1.bar(avg_by_year['year'], avg_by_year['comb08'])
ax1.set(title='Avg By Year', ylabel='MPG', xlabel='Year');
```

```python
fig, ax1 = plt.subplots()
ax1.bar(avg_by_year['year'], avg_by_year['comb08'])
ax1.set_xticks(np.arange(2000, 2022, 2))
ax1.set(title='Avg By Year', ylabel='MPG', xlabel='Year');
```

```python
fig, ax1 = plt.subplots()
ax1.barh(avg_by_year['year'], avg_by_year['comb08'])
ax1.set(title='Avg By Year', xlabel='MPG', ylabel='Year');
```

```python
fig, ax1 = plt.subplots()
ax1.barh(avg_by_year['year'], avg_by_year['comb08'])
ax1.set_yticks(np.arange(2000, 2022, 2))
ax1.set(title='Avg By Year', xlabel='MPG', ylabel='Year');
```

```python
fig, ax1 = plt.subplots()
ax1.scatter(x=df['fuelCost08'], y=df['displ'], alpha=0.5, c=df['cylinders'])
ax1.set(ylabel='Displacement', xlabel='Cost');
```

```python
fig, ax1 = plt.subplots(figsize=(7,5))
ax1.scatter(x=df['fuelCost08'], y=df['displ'], alpha=0.5, c=df['cylinders'])
ax1.set_ylabel('Displacement', size=12)
ax1.set_xlabel('Cost', size=12)
ax1.xaxis.set_major_formatter('${x:,.0f}')
ax1.tick_params(axis='x', labelrotation=45, labelsize=12)
ax1.tick_params(axis='y', labelsize=12)
ax1.axvline(3500, color='black', linestyle=':')
ax1.annotate('Target of $3500', xy=(3500, 2), size=14)
ax1.grid(True)
```
