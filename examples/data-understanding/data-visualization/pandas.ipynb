```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
import os
```

```python
dataset = os.path.join('datasets', 'EPA_fuel_economy.csv')
df = pd.read_csv(dataset)
df
```

```python
df.info()
```

```python
df['comb08'].plot(kind='hist')
```

```python
df['comb08'].plot(kind='box')
```

```python
df['comb08'].plot(kind='kde')
```

```python
avg_by_year = df.groupby(['year']).agg({'highway08': 'mean', 'city08': 'mean', 'comb08': 'mean'})
avg_by_year
```

```python
avg_by_year.plot(kind='box')
plt.grid(True)
```

```python
avg_by_year.plot(kind='line')
plt.xticks(np.arange(2000,2022,2));
```

```python
avg_by_year.plot(kind='line', figsize=(9,5), xticks=np.arange(2000,2022,2), ylim=(10,32), rot=45)
```

```python
avg_by_year.plot(kind='bar', figsize=(14, 4))
```

```python
avg_by_year.plot(kind='barh', figsize=(4, 14))
```

```python
avg_by_year.plot(kind='area')
```

```python
fig, ax = plt.subplots(figsize=(9,4))
avg_by_year.plot(kind='area', ax=ax)
```

```python
fig, ax1 = plt.subplots(figsize=(9,4))
avg_by_year.plot(kind='area', ax=ax1, stacked=False)
ax1.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:0.0f}'))
ax1.set(ylabel='MPG', title='Avg MPG by Year');
```
